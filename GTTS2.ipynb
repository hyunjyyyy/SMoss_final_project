```python
!pip install google-cloud-texttospeech

from google.colab import files

# 파일 업로드
uploaded = files.upload()

# 업로드한 파일 이름을 확인합니다.
for filename in uploaded.keys():
    print(f'Uploaded file: {filename}')

import os

# 업로드한 파일 이름을 경로로 설정합니다.
key_file = list(uploaded.keys())[0]
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = f"/content/{key_file}"

from google.cloud import texttospeech

# 클라이언트 생성
client = texttospeech.TextToSpeechClient()

# TTS 요청을 클래스에 따라 설정하는 함수
def generate_tts_text(class_number):
    if class_number == 1:
        return "이것은 플라스틱 쓰레기입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 2:
        return "이것은 종이 쓰레기입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 3:
        return "이것은 금속 쓰레기입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 4:
        return "이것은 유리 쓰레기입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 5:
        return "이것은 플라스틱 병입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 6:
        return "이것은 종이팩입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 7:
        return "이것은 일반 쓰레기입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 8:
        return "이것은 음식물 쓰레기입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 9:
        return "이것은 옷입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 10:
        return "이것은 전자제품입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 11:
        return "이것은 종이팩입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 12:
        return "이것은 종이팩입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 13:
        return "이것은 종이팩입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 14:
        return "이것은 종이팩입니다. 재활용 방법은 다음과 같습니다."
    elif class_number == 15:
        return "이것은 종이팩입니다. 재활용 방법은 다음과 같습니다."
    else:
        return "알 수 없는 쓰레기입니다."

# 클래스 번호 설정
class_number = 1  # 예시로 클래스 번호를 1로 설정합니다. 실제 객체 인식 결과에 따라 이 값을 바꿔야 합니다.

# TTS 요청 생성
text_to_speak = generate_tts_text(class_number)
synthesis_input = texttospeech.SynthesisInput(text=text_to_speak)

# TTS 요청 설정
voice = texttospeech.VoiceSelectionParams(
    language_code="ko-KR",  # 한국어 코드
    ssml_gender=texttospeech.SsmlVoiceGender.NEUTRAL
)
audio_config = texttospeech.AudioConfig(audio_encoding=texttospeech.AudioEncoding.MP3)

# 요청 보내기 및 응답 받기
response = client.synthesize_speech(input=synthesis_input, voice=voice, audio_config=audio_config)

# 응답 저장
output_path = "/content/output.mp3"
with open(output_path, "wb") as out:
    out.write(response.audio_content)
    print(f'Audio content written to file "{output_path}"')

# 생성된 오디오 파일 다운로드
files.download(output_path)
```
