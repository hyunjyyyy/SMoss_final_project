```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torch.utils.data import Dataset, DataLoader, random_split, WeightedRandomSampler
from PIL import Image
import pandas as pd
import os
import json
import matplotlib.pyplot as plt
from torch.optim.lr_scheduler import ReduceLROnPlateau

# GPU 사용 가능 확인 및 장치 설정
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 데이터 프레임 생성 (이미지와 라벨 정보 포함)
image_root = '/content/SMoss_final_project/Sample/01.원천데이터/'
label_root = '/content/SMoss_final_project/Sample/02.라벨링데이터/'

data = []

for dirpath, _, filenames in os.walk(image_root):
    for filename in filenames:
        if filename.endswith('.jpg'):
            image_path = os.path.join(dirpath, filename)
            label_path = os.path.join(label_root, os.path.relpath(dirpath, image_root), filename.replace('.jpg', '.json'))

            with open(label_path, 'r') as f:
                label_data = json.load(f)
                image_info = label_data['IMAGE_INFO']
                annotation_info = label_data['ANNOTATION_INFO'][0]
                label = annotation_info['CLASS']

            data.append({
                'image_path': image_path,
                'label': label
            })

df = pd.DataFrame(data)

# 클래스 인덱스를 문자열 라벨에서 정수로 변환하기 위해 고유 라벨 목록 생성
label_to_index = {label: idx for idx, label in enumerate(df['label'].unique())}
df['label'] = df['label'].map(label_to_index)

# 클래스별 데이터 수 계산
class_counts = df['label'].value_counts().to_dict()

# 가장 많이 있는 클래스의 데이터 수 찾기
max_count = max(class_counts.values())

# 데이터 증강 비율 계산
class_weights = {label: max_count / count for label, count in class_counts.items()}

# 데이터 증강을 위한 전처리 함수 정의
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),  # 추가: 수직 뒤집기
    transforms.RandomRotation(20),  # 강화된 회전
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.3),  # 강화된 색상 조정
    transforms.RandomGrayscale(p=0.1),  # 일부 이미지를 무작위로 흑백으로 변환
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

test_transform = transforms.Compose([
    transforms.Resize((224, 224)),  # ResNet-50 입력 크기에 맞춤
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # ImageNet 데이터셋의 평균과 표준편차 사용
])

# 사용자 정의 데이터셋 클래스 정의
class CustomDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform
        self.weights = self.calculate_weights()

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_name = self.dataframe.iloc[idx, 0]
        image = Image.open(img_name).convert('RGB')
        label = self.dataframe.iloc[idx, 1]

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label)  # label을 텐서로 변환하여 반환

    def calculate_weights(self):
        weights = self.dataframe['label'].map(class_weights).values
        return weights

# 데이터셋 생성
dataset = CustomDataset(df, transform=train_transform)

# 데이터셋을 학습과 검증으로 분리
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# WeightedRandomSampler를 사용하여 클래스 비율에 맞게 샘플링
train_sampler = WeightedRandomSampler(weights=dataset.calculate_weights()[:train_size], num_samples=train_size, replacement=True)

# DataLoader 생성
batch_size = 32  # 배치 크기 변경
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# 모델 정의 함수 (densenet 모델 추가)
def get_model(model_name, num_classes):
    if model_name == 'resnet18':
        model = models.resnet18(pretrained=True)
        num_ftrs = model.fc.in_features
        model.fc = nn.Linear(num_ftrs, num_classes)
    elif model_name == 'resnet50':
        model = models.resnet50(pretrained=True)
        num_ftrs = model.fc.in_features
        model.fc = nn.Linear(num_ftrs, num_classes)
    elif model_name == 'resnet101':
        model = models.resnet101(pretrained=True)
        num_ftrs = model.fc.in_features
        model.fc = nn.Linear(num_ftrs, num_classes)
    elif model_name == 'vgg16':
        model = models.vgg16(pretrained=True)
        num_ftrs = model.classifier[6].in_features
        model.classifier[6] = nn.Linear(num_ftrs, num_classes)
    elif model_name == 'densenet121':
        model = models.densenet121(pretrained=True)
        num_ftrs = model.classifier.in_features
        model.classifier = nn.Linear(num_ftrs, num_classes)
    else:
        raise ValueError(f'Unknown model name {model_name}')

    return model

# 손실 함수 및 최적화 기법 정의
def get_optimizer_scheduler(model):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.0001, weight_decay=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3, verbose=True)
    return criterion, optimizer, scheduler

# 학습 함수 정의 (학습 및 검증 손실/정확도 기록)
def train_model(model, train_dataloader, val_dataloader, criterion, optimizer, scheduler, num_epochs=50, patience=5):
    train_loss_history = []
    val_loss_history = []
    train_acc_history = []
    val_acc_history = []

    best_val_loss = float('inf')
    epochs_no_improve = 0
    best_model_wts = None

    for epoch in range(num_epochs):
        print(f'에폭 {epoch+1}/{num_epochs}')
        print('-' * 10)

        # 학습 단계
        model.train()
        running_loss = 0.0
        correct_predictions = 0
        for inputs, labels in train_dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            correct_predictions += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = correct_predictions.double() / len(train_dataloader.dataset)

        train_loss_history.append(epoch_loss)
        train_acc_history.append(epoch_acc.item())

        print(f"학습 손실: {epoch_loss:.4f} 정확도: {epoch_acc:.4f}")

        # 검증 단계
        model.eval()
        running_loss = 0.0
        correct_predictions = 0
        with torch.no_grad():
            for inputs, labels in val_dataloader:
                inputs, labels = inputs.to(device), labels.to(device)

                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                correct_predictions += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(val_dataloader.dataset)
        epoch_acc = correct_predictions.double() / len(val_dataloader.dataset)

        val_loss_history.append(epoch_loss)
        val_acc_history.append(epoch_acc.item())

        print(f"검증 손실: {epoch_loss:.4f} 정확도: {epoch_acc:.4f}")

        # ReduceLROnPlateau 스케줄러 step 호출
        scheduler.step(epoch_loss)

        # 조기 종료 확인 및 최적 모델 저장
        if epoch_loss < best_val_loss:
            best_val_loss = epoch_loss
            epochs_no_improve = 0
            best_model_wts = model.state_dict()
            save_model(model, f"best_{model_name}")  # 최적 모델 저장
        else:
            epochs_no_improve += 1

        if epochs_no_improve == patience:
            print(f'조기 종료! 검증 손실이 {patience} 에폭 동안 개선되지 않아 에폭 {epoch+1} 에서 학습 중지.')
            break

    # 최적 모델 상태로 복원
    if best_model_wts is not None:
        model.load_state_dict(best_model_wts)

    # 학습 및 검증 손실/정확도 시각화
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(train_loss_history, label='학습 손실')
    plt.plot(val_loss_history, label='검증 손실')
    plt.xlabel('에폭')
    plt.ylabel('손실')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(train_acc_history, label='학습 정확도')
    plt.plot(val_acc_history, label='검증 정확도')
    plt.xlabel('에폭')
    plt.ylabel('정확도')
    plt.legend()
    plt.tight_layout()
    plt.show()

    return model

# 모델 저장 함수
def save_model(model, model_name, save_dir='./saved_models'):
    os.makedirs(save_dir, exist_ok=True)
    save_path = os.path.join(save_dir, f'{model_name}.pth')
    torch.save(model.state_dict(), save_path)
    print(f'{model_name} 저장 완료: {save_path}')

# 모델 로드 함수
def load_model(model_name, model_class, num_classes, save_dir='./saved_models'):
    model = get_model(model_class, num_classes).to(device)
    save_path = os.path.join(save_dir, f'{model_name}.pth')
    model.load_state_dict(torch.load(save_path, map_location=device))
    print(f'{model_name} 로드 완료: {save_path}')
    return model

# 모델 이름과 클래스 수 설정
model_names = ['resnet18', 'resnet50', 'densenet121']  # 여기에 densenet 모델 추가 가능
num_classes = len(label_to_index)

# 모델 학습 및 저장
trained_models = {}
for model_name in model_names:
    print(f'현재 학습 중인 모델: {model_name}')
    model = get_model(model_name, num_classes).to(device)
    criterion, optimizer, scheduler = get_optimizer_scheduler(model)
    trained_model = train_model(model, train_dataloader, val_dataloader, criterion, optimizer, scheduler)
    trained_models[model_name] = trained_model

# 모델 로드 예시
loaded_models = {}
for model_name in model_names:
    loaded_model = load_model(f"best_{model_name}", model_name, num_classes)
    loaded_models[model_name] = loaded_model

# 앙상블 예측 생성
def ensemble_predictions(models, dataloader):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    predictions = []

    num_classes = len(label_to_index)  # 클래스 수를 데이터셋에 맞게 설정

    for inputs, _ in dataloader:
        inputs = inputs.to(device)

        # 각 모델에 대해 softmax 적용 후 평균 계산
        avg_outputs = torch.zeros(inputs.size(0), num_classes).to(device)
        for model in models:
            model.eval()
            with torch.no_grad():
                outputs = torch.softmax(model(inputs), dim=1)
                avg_outputs += outputs / len(models)  # 모델 수에 맞게 평균을 더함

        _, preds = torch.max(avg_outputs, 1)
        predictions.extend(preds.cpu().numpy())

    return predictions

# 검증 데이터셋에 대한 각 모델의 예측 생성
ensemble_models = [loaded_models[model_name] for model_name in model_names]
val_predictions = ensemble_predictions(ensemble_models, val_dataloader)

# 예측 평가
val_labels = [label.item() for _, label in val_dataset]
accuracy = sum(1 for x, y in zip(val_predictions, val_labels) if x == y) / len(val_labels)
print(f'앙상블 검증 정확도: {accuracy:.4f}')

# 이미지 분류 함수
def classify_image(image_path, models):
    # 이미지 열기
    image = Image.open(image_path).convert('RGB')

    # 이미지 전처리
    image = test_transform(image).unsqueeze(0)  # 배치 차원을 추가

    # GPU로 이미지 옮기기
    image = image.to(device)

    # 모델로 예측
    avg_outputs = torch.zeros(1, num_classes).to(device)
    with torch.no_grad():
        for model in models:
            model.eval()
            outputs = torch.softmax(model(image), dim=1)
            avg_outputs += outputs / len(models)  # 모델 수에 맞게 평균을 더함

    _, preds = torch.max(avg_outputs, 1)
    predicted_label = list(label_to_index.keys())[preds.item()]  # 인덱스를 실제 라벨로 변환
    return predicted_label

# 테스트할 이미지 경로
test_image_path = '/content/스티로폼.jpg'

# 이미지 분류
predicted_label = classify_image(test_image_path, ensemble_models)
print(f'The predicted label for the image is: {predicted_label}')
```

```python
import torch
import torch.nn as nn
from torchvision import models, transforms
from PIL import Image

# 클래스 인덱스를 문자열 라벨로 변환하기 위한 사전
index_to_label = {idx: label for label, idx in label_to_index.items()}

# 이미지 전처리를 위한 변환 정의
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # ResNet-50 입력 크기에 맞춤
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # ImageNet 데이터셋의 평균과 표준편차 사용
])

# 사전 학습된 ResNet-50 모델 불러오기
model = models.resnet50(pretrained=False)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, len(index_to_label))  # 클래스 수는 학습된 모델과 동일하게 설정

# 저장된 모델 상태 불러오기
model_state_dict = torch.load('/content/saved_models/best_resnet50.pth')
new_model_state_dict = model.state_dict()

# 필요한 파라미터만 불러오기
for name, param in model_state_dict.items():
    if name in new_model_state_dict:
        new_model_state_dict[name].copy_(param)

model.load_state_dict(new_model_state_dict)
model.eval()  # 평가 모드로 설정

# GPU 사용 가능 여부 확인 및 모델을 GPU로 옮기기
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = model.to(device)

def classify_image(image_path):
    # 이미지 열기
    image = Image.open(image_path).convert('RGB')

    # 이미지 전처리
    image = transform(image).unsqueeze(0)  # 배치 차원을 추가

    # GPU로 이미지 옮기기
    image = image.to(device)

    # 모델로 예측
    with torch.no_grad():
        outputs = model(image)
        _, preds = torch.max(outputs, 1)

    # 결과 반환
    predicted_label = index_to_label[preds.item()]
    return predicted_label

# 테스트할 이미지 경로
test_image_path = '/content/페트병2.jpg'

# 이미지 분류
predicted_label = classify_image(test_image_path)
print(f'The predicted label for the image is: {predicted_label}')


```

```python

# 이미지 분류 함수 (상위 3개의 클래스와 확률 반환)
def classify_image(image_path, models):
    # 이미지 열기
    image = Image.open(image_path).convert('RGB')

    # 이미지 전처리
    image = test_transform(image).unsqueeze(0)  # 배치 차원을 추가

    # GPU로 이미지 옮기기
    image = image.to(device)

    # 모델로 예측
    avg_outputs = torch.zeros(1, num_classes).to(device)
    with torch.no_grad():
        for model in models:
            model.eval()
            outputs = torch.softmax(model(image), dim=1)
            avg_outputs += outputs / len(models)  # 모델 수에 맞게 평균을 더함

    topk_probs, topk_indices = torch.topk(avg_outputs, 3, dim=1)
    topk_probs = topk_probs.cpu().numpy().flatten()
    topk_indices = topk_indices.cpu().numpy().flatten()

    topk_labels = [list(label_to_index.keys())[index] for index in topk_indices]
    return list(zip(topk_labels, topk_probs))

# 테스트할 이미지 경로
test_image_path = '/content/페트병2.jpg'

# 이미지 분류
top3_predictions = classify_image(test_image_path, ensemble_models)
print(f'Top 3 predicted labels and probabilities for the image: {top3_predictions}')
```
