```python
!pip install google-cloud-texttospeech

from google.colab import files

# 파일 업로드
uploaded = files.upload()

# 업로드한 파일 이름을 확인합니다.
for filename in uploaded.keys():
    print(f'Uploaded file: {filename}')

import os

# 업로드한 파일 이름을 경로로 설정합니다.
key_file = list(uploaded.keys())[0]
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = f"/content/{key_file}"

from google.cloud import texttospeech

# 클라이언트 생성
client = texttospeech.TextToSpeechClient()

# TTS 요청을 클래스에 따라 설정하는 함수
def generate_tts_text(predicted_label):
    if predicted_label == "플라스틱":
        return "이것은 플라스틱 쓰레기입니다. 1. 내용물을 비우고 물로 헹구는 등 이물질을 제거하여 배출한다. 2. 부착상표, 부속품 등 본체와 다른 재질은 제거한 후 배출한다. "
    elif predicted_label == "종이":
        return "이것은 종이 쓰레기입니다. 신문일 경우 물기에 젖지 않도록 하고, 반듯하게 펴서 차곡차곡 쌓은 후 흩날리지 않도록 끈 등으로 묶어서 배출한다. 노트일 경우 스프링 등 종이류와 다른 재질은 제거한 후 배출한다. 상자류일 경우 테이프 등 종이류와 다른 재질은 제거한 후 배출한다."
    elif predicted_label == "금속캔":
        return "이것은 금속캔입니다. 그러므로 1. 내용물을 비우고 물로 헹구는 등 이물질을 제거하여 배출합니다. 2. 담배꽁초 등 이물질을 넣지 않고 배출합니다. 3. 플라스틱 뚜껑 등 금속캔과 다른 재질은 제거한 후 배출합니다."
    elif predicted_label == "유리병":
        return "이것은 유리병입니다. 1. 내용물을 비우고 물로 헹구는 등 이물질을 제거하여 배출한다. 2. 담배꽁초 등 이물질을 넣지 않고 배출한다. 3. 유리병이 깨지지 않도록 주의하여 배출한다. 4. 소주, 맥주 등 빈용기보증금 대상 유리병은 소매점 등으로 반납하여 보증금 환급한다"
    elif predicted_label == "페트병":
        return "이것은 페트병입니다. 1. 내용물을 비우고 물로 헹구는 등 이물질을 제거하여 배출한다. 2. 부착상표, 부속품 등 본체와 다른 재질은 제거한 후 배출한다"
    elif predicted_label == "스티로폼":
        return "이것은 스티로폼입니다. 1. 내용물을 비우고 물로 헹구는 등 이물질을 제거하여 배출한다. 2. 부착상표 등 스티로폼과 다른 재질은 제거한 후 배출한다. 3. TV 등 전자제품 구입 시 완충재로 사용되는 발포합성수지 포장재는 가급적 구입처로 반납한다"
    elif predicted_label == "비닐":
        return "이것은 비닐입니다. 1. 내용물을 비우고 물로 헹구는 등 이물질을 제거하여 배출한다. 2. 흩날리지 않도록 봉투에 담아 배출한다."
    elif predicted_label == "형광등":
        return "이것은 형광등입니다. 1. 폐형광등은 유해물질을 포함하고 있으므로 깨지지 않도록 주의하여 폐형광등 전용수거함에 배출한다. 2. 깨진 폐형광등은 신문지 등으로 감싸 사람이 찔리거나 베이지 않도록 하여 종량제 봉투에 담아 배출한다. "
    elif predicted_label == "건전지":
        return "이것은 건전지입니다. 1. 전지를 제품에서 분리하여 배출한다. 2. 주요거점(동 주민센터·아파트의 폐형광등·폐건전지 일체함 또는 편의점, 아파트 동별 우편함 등)의 전용수거함에 배출한다 "
    else:
        return "알 수 없는 쓰레기입니다."


# TTS 요청 생성
text_to_speak = generate_tts_text(predicted_label)
synthesis_input = texttospeech.SynthesisInput(text=text_to_speak)

# TTS 요청 설정
voice = texttospeech.VoiceSelectionParams(
    language_code="ko-KR",  # 한국어 코드
    ssml_gender=texttospeech.SsmlVoiceGender.NEUTRAL
)
audio_config = texttospeech.AudioConfig(audio_encoding=texttospeech.AudioEncoding.MP3)

# 요청 보내기 및 응답 받기
response = client.synthesize_speech(input=synthesis_input, voice=voice, audio_config=audio_config)

# 응답 저장
output_path = "/content/output.mp3"
with open(output_path, "wb") as out:
    out.write(response.audio_content)
    print(f'Audio content written to file "{output_path}"')

# 생성된 오디오 파일 다운로드
files.download(output_path)
```
